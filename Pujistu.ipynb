```python
import cv2
import os

def compare_images(query_image_path, directory_path, threshold):
    # Read query image
    query_image = cv2.imread(query_image_path, 0)
    query_image = cv2.resize(query_image, (500, 500))

    # Create SIFT object
    sift = cv2.SIFT_create()

    # Detect keypoints and compute descriptors for query image
    kp_query, desc_query = sift.detectAndCompute(query_image, None)

    # Create BFMatcher object
    bf = cv2.BFMatcher()

    similarity_scores = {}

    # Loop through images in directory
    for filename in os.listdir(directory_path):
        image_path = os.path.join(directory_path, filename)
        print("Directory and image_path: ", image_path)
        if image_path.endswith(".jpg"):
            # Read directory image
            img = cv2.imread(image_path, 0)
            img = cv2.resize(img, (500, 500))

            # Detect keypoints and compute descriptors for directory image
            kp, desc = sift.detectAndCompute(img, None)

            # Match descriptors using BFMatcher
            matches = bf.knnMatch(desc_query, desc, k=2)

            # Apply ratio test to get good matches
            good_matches = []
            if len(matches) >= 2:
                for m, n in matches:
                    if m.distance < 0.75 * n.distance:
                        good_matches.append(m)

            # Compute similarity score based on number of good matches
            similarity_score = len(good_matches) / len(matches)

            # Add similarity score and filename to dictionary
            similarity_scores[filename] = similarity_score

            # Print similarity score and filename
            print("similarity_score: ", similarity_score, " filename: ", filename)

    # Filter similarity scores based on threshold
    similarity_scores = {k: v for k, v in similarity_scores.items() if v >= threshold}

    # Sort similarity scores by descending order of score
    similarity_scores = dict(sorted(similarity_scores.items(), key=lambda item: item[1], reverse=True))

    return similarity_scores


query_image_path = 'images/1.jpg'
directory_path = 'images/'
threshold = 0.2
similarity_scores = compare_images(query_image_path, directory_path, threshold)
print(similarity_scores)
```
